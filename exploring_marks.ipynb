```python
df['Math_PassStatus'] = np.where(df['math score']<40,'Failed','Passed')
df.Math_PassStatus.value_counts()
```

```python
df['OverallPassStatus']=np.where((df['reading score']<40 and df['reading score']<40 and df['math score']<40),'Failed','Passed')
```
